# Heteroscedastic

```python
hetero=pd.read_csv('heteroscedastic.csv')
plt.scatter(hetero['x'],hetero['y'])
plt.show()
```

```python
train, test = train_test_split(hetero, test_size=0.2,random_state=0)

fig, ax = plt.subplots(1,2, figsize=(15, 5))
ax[0].scatter(train['x'],train['y'])
ax[0].set_title("Train Data")

ax[1].scatter(test['x'],test['y'])
ax[1].set_title("Test Data")
plt.show()
```

# Monotonic

```python
monotonic=pd.read_csv('monotonic.csv')
plt.scatter(monotonic['x'],monotonic['y'])
plt.show()
```

```python
train, test = train_test_split(monotonic, test_size=0.2,random_state=0)

fig, ax = plt.subplots(1,2, figsize=(15, 5))
ax[0].scatter(train['x'],train['y'])
ax[0].set_title("Train Data")

ax[1].scatter(test['x'],test['y'])
ax[1].set_title("Test Data")
plt.show()
```
